```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading the file

```python
iris_df = pd.read_excel('iris.xls')
display(iris_df.head())
```

Columns

```python
iris_df.columns
```

Mean

```python
display(iris_df.mean(numeric_only=True))
```

Checking for null values

```python
display(iris_df.isnull().sum())
```

Viusualisation

```python
numerical_cols = iris_df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=iris_df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numerical_cols = iris_df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=iris_df, x='Classification', y=col)
    plt.title(f'Distribution of {col} across Classification')
    plt.xlabel('Classification')
    plt.ylabel(col)
    plt.show()
```

```python
sns.pairplot(iris_df, hue='Classification')
plt.suptitle('Pair Plot of Iris Dataset by Classification', y=1.02)
plt.show()
```
